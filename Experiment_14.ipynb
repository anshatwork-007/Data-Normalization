## Title: Data Normalization


```python
import pandas as pd
import numpy as np
data = {
    "Product": ['laptop', 'mobile', 'tablet', 'headphone', 'smartwatch', 'camera'],
    'Price': [55000,20000,30000,2000,8000,45000],
    'Units_Sold': [25,60,35,80,50,15],
    'Discount': [10,15,12,20,18,8]
}

df = pd.DataFrame(data)

print("Original Dataset")
print(df)
```

```python
# 1 Min-Max Normalization

df['Price_Min-Max'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())

print("\nMin-Max Normalized Price")
print(df[['Product','Price','Price_Min-Max']])
```

```python
# Normalize multiple columns
cols = ['Price', 'Units_Sold', 'Discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNormalized Dataset")
print(df_norm)
```

```python
# Z-score Normalization

df['Units_Z-score'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
print("\nZ-score Normalized Units_Sold")
print(df[['Product','Units_Sold','Units_Z-score']])
```

```python
# Decimal Scaling

df['Price_Decimals'] = df['Price']/100000
print("\nDecimal Scaled Price")
print(df[['Product','Price','Price_Decimals']])
```

```python
# Turning Categorical Variables into quantitative variables
import pandas as pd

data = {
    'Order_ID': [101, 102, 103, 104, 105, 106],
    'Customer_Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'Payment_Method': ['UPI', 'Credit Card', 'Debit Card', 'COD', 'UPI', 'Debit Card'],
    'Product_Category': ['Electronics', 'Clothing', 'Home', 'Electronics', 'Beauty', 'Electronics'],
    'City': ['Pune', 'Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Hyderabad'],
    'Order_Value': [4500, 2300, 3200, 5200, 1800, 4100]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the output
print(df)
```

```python
# Method 1:Label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender_Label'] = le.fit_transform(df['Customer_Gender'])

print(df)
```

```python
# Method 2 Gate dummies
df_encoded = pd.get_dummies(df, columns=['Payment_Method'])

print(df_encoded)
```

```python
# Encoding Multiple columns

df_multi = pd.get_dummies(df, columns=['Product_Category','City'])

print(df_multi)
```

```python
# Method 3: Dummy encoding( Drop First)

df_dummy = pd.get_dummies(df, columns=['Product_Category'], drop_first=True)

print(df_dummy)
```
